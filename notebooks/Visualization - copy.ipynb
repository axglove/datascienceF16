```python
columns = """-- The attributes are (dontated by Riccardo Leardi, 
    riclea@anchem.unige.it )
    1) Alcohol
    2) Malic acid
    3) Ash
    4) Alcalinity of ash  
    5) Magnesium
    6) Total phenols
    7) Flavanoids
    8) Nonflavanoid phenols
    9) Proanthocyanins
    10)Color intensity
    11)Hue
    12)OD280/OD315 of diluted wines
    13)Proline"""

columns = columns.split('\n')
columns = [i.strip() for i in columns][2:]

# how does this work? - splits at the ) and takes the second element, then removes spaces and lowercases
columns = [i.split(')')[1].strip().lower() for i in columns]

# what does this do? - adds a class column at the beginning of the list
cols = ['class']
cols.extend(columns)

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
wine = pd.read_csv(url, header=None)
wine.columns = cols
wine.head()
```

```python
winegroups = wine.groupby('class')
print type(winegroups)
#print winegroups.head()

winegroups.describe()
```

```python
winegroups.sum()
```

```python
print winegroups.alcohol.mean()
print winegroups[['alcohol', 'ash']].median()

#print winegroups.alcohol.median()
#print winegroups.agg({'alcohol': 'median'})

print winegroups.alcohol.nunique()
print winegroups.agg({'alcohol': lambda alcohol: alcohol.nunique() / alcohol.count()})
```

```python
#print len(wine)
#print wine.sort('alcohol').head(40)

#if you remove the count function, aggfunc defaults to mean
#what is the count function counting? # of ash values?
pivoted_wine = wine.pivot_table(values='ash',
                 columns='class',
                 index='alcohol',
                 aggfunc='count')    

print pivoted_wine.head()
#print len(pivoted_wine)

#not sure what reset_index does here
pivoted_wine_reset = pivoted_wine.reset_index()
#print len(pivoted_wine_reset)
print pivoted_wine_reset.head()
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
wine = pd.read_csv(url, header=None)
wine.columns = cols

#create a new column with the rounded alcohol content
wine['rounded alc'] = wine['alcohol'].round()
wine.head()

#groupby on the new column
alcgroups = wine.groupby('rounded alc')

alcgroups.describe()
alcgroups.mean()
alcgroups.std()

#ash content increases
#alcalinity of ash decreases
#color intensity increases?
#proline increases
```

```python
pivoted_wine_alex = wine.pivot_table(values='alcohol',
                 columns='class',
                 index='rounded alc',
                 aggfunc='count')    

print pivoted_wine_alex.head()


pivoted_wine_alex_reset = pivoted_wine_alex.reset_index()

print pivoted_wine_alex_reset.head()

print pd.melt(pivoted_wine_reset,
              id_vars=['alcohol'],
              value_vars=[1, 2, 3],
              value_name='count').dropna().head()

print pd.melt(pivoted_wine_alex_reset,
              id_vars=['class'],
              value_vars=['rounded alc'],
              value_name='count').dropna().head()
```
